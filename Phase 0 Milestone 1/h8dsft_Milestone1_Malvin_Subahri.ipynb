### `Analisa Central Tendency`

Untuk mengetahui Google Ads apa yang akan dipakai nantinya, ada baiknya kita mengetahui rata - rata biaya dan rata - rata biaya per hari untuk Google Ads jenis VIDEO & IMAGE

````python
video_mean= ads[ads['ad_type']=='VIDEO'][['ad_type','spend_usd']].groupby('ad_type').mean()
text_mean= ads[ads['ad_type']=='TEXT'][['ad_type','spend_usd']].groupby('ad_type').mean()

print('Rata - Rata Biaya Untuk Setiap Jenis VIDEO : {}'.format(np.round(video_mean.spend_usd.mean())))
print('Rata - Rata Biaya Untuk Setiap Jenis TEXT : {}'.format(np.round(text_mean.spend_usd.mean())))
````

Rata - rata biaya untuk setiap jenis TEXT lebih murah daripada VIDEO

````python
daily_video = ads[ads['ad_type']=='VIDEO'][['date_range_start','spend_usd']].groupby('date_range_start').sum()
daily_text = ads[ads['ad_type']=='TEXT'][['date_range_start','spend_usd']].groupby('date_range_start').sum()

print('Rata - Rata Biaya Yang Dikeluarkan Untuk Jenis VIDEO / Hari : {}'.format(np.round(daily_video.spend_usd.mean())))
print('Rata - Rata Biaya Yang Dikeluarkan Untuk Jenis TEXT / Hari : {}'.format(np.round(daily_text.spend_usd.mean())))
````

Rata - rata biaya yang dikeluarkan per hari untuk VIDEO lebih murah daripada TEXT

````python
plt.figure(figsize=(25,5))

sns.distplot(daily_video,color='blue', label='Average VIDEO Cost a Day', bins = 25)
sns.distplot(daily_text,color='red', label='Average TEXT Cost a Day', bins = 25)

plt.axvline(video_mean.spend_usd.mean(), color='green', label='VIDEO Mean', linewidth=2, linestyle = 'dashed')
plt.axvline(text_mean.spend_usd.mean(), color='yellow', label='TEXT Mean', linewidth=4, linestyle = 'dashed')

plt.legend ()
````

Rata - rata biaya untuk setiap jenis TEXT lebih murah daripada VIDEO

Rata - rata biaya yang dikeluarkan per hari untuk VIDEO lebih murah daripada TEXT

Dari sini kita bisa melihat bahwa ads jenis TEXT lebih banyak daripada ads jenis VIDEO karena biarpun rata - rata biaya jenis TEXT lebih murah tetapi biaya yang dikeluarkan per hari untuk ads jenis TEXT lebih besar daripada ads jenis VIDEO

### `Hipotesis Testing`

Untuk menentukan Google Ads jenis apakah yang akan dipakai, kita menggunakan `Two Samples Independent Two Tailed Hypothesis Testing` untuk mengecek apakah rata-rata biaya yang dikeluarkan / hari untuk jenis iklan VIDEO & TEXT berbeda secara signifikan atau tidak

Hipotesis kita dalam kasus ini:

H0: Î¼_video = Î¼_text

H1: Î¼_video != Î¼_text

````python
t_stat, p_val = stats.ttest_ind(daily_video, daily_text)
print('P-value:',p_val[0]) #the p-value isn't divided by 2 since the output is two-sided p-value
print('t-statistics:',t_stat[0])
````

````python
video_pop = np.random.normal(daily_video.spend_usd.mean(),daily_video.spend_usd.std(),10000)
text_pop = np.random.normal(daily_text.spend_usd.mean(),daily_text.spend_usd.std(),10000)

ci = stats.norm.interval(0.95, daily_video.spend_usd.mean(), daily_video.spend_usd.std())
plt.figure(figsize=(16,5))
sns.distplot(video_pop, label='Average VIDEO Cost a Day',color='blue')
sns.distplot(text_pop, label='Average TEXT Cost a Day',color='red')

plt.axvline(daily_video.spend_usd.mean(), color='blue', linewidth=2, label='VIDEO Mean')
plt.axvline(daily_text.spend_usd.mean(), color='red',  linewidth=2, label='TEXT Mean')

plt.axvline(ci[1], color='green', linestyle='dashed', linewidth=2, label='Confidence Threshold of 95%')
plt.axvline(ci[0], color='green', linestyle='dashed', linewidth=2)

plt.axvline(video_pop.mean()+t_stat[0]*video_pop.std(), color='black', linestyle='dashed', linewidth=2, label = 'Alternative Hypothesis')
plt.axvline(video_pop.mean()-t_stat[0]*video_pop.std(), color='black', linestyle='dashed', linewidth=2)

plt.legend()
````

Berdasarkan hasil di atas, kita dapat menyimpulkan bahwa kita gagal untuk merejek H0 / hipotesis nol di mana diantara VIDEO dan TEXT tidak berbeda secara signifikan dalam hal rata - rata biaya per hari

Karena rata - rata biaya yang dikeluarkan / hari untuk jenis iklan VIDEO dan TEXT tidak berbeda secara signifikan, kita bisa memilih di antara kedua jenis itu, namun jika dilihat dari popularitas, jenis TEXT merupakan jenis iklan yang lebih popular dan lebih banyak dipakai di Google Ads
