# Eksplorasi Data

Bagian ini berisi explorasi data pada dataset diatas dengan menggunakan query, grouping, visualisasi sederhana, dan lain sebagainya.

```python
poke
```

```python
poke.describe ().T
```

```python
poke ['Type 1'].unique ()
```

Dari type pokemon di atas, maka type pokemon yang akan kita ambil untuk mewakili militer dunia pokemon adalah sbb:
- Angkatan darat = ground
- Angkatan laut = water
- Angkatan udara = flying

```python
poke [(poke ['Generation'] >= 1) &
    (poke ['Type 1'] == 'Ground') &
    (poke ['Legendary'] == True)
][['Name', 'Generation', 'Total']]
# Dari hasil query, pokemon legendary type ground terkuat yang cocok menjadi komandan pasukan angkatan darat adalah 'GroudonPrimal Groudon'
```

```python
poke [(poke ['Generation'] >= 1) &
    (poke ['Type 1'] == 'Water') &
    (poke ['Legendary'] == True)
][['Name', 'Generation', 'Total']]
# Dari hasil query, pokemon legendary type water terkuat yang cocok menjadi komandan pasukan angkatan laut adalah 'KyogrePrimal Kyogre' 
```

```python
poke [(poke ['Generation'] >= 1) &
    (poke ['Type 1'] == 'Flying') &
    (poke ['Legendary'] == True)
][['Name', 'Generation', 'Total']]
# Dari hasil query, pokemon legendary type flying terkuat yang cocok menjadi komandan pasukan angkatan udara adalah: 
# 'TornadusIncarnate Forme / 'TornadusTherian Forme' 
```

```python
poke [(poke ['Generation'] >= 1) &
    (poke ['Type 1'] == 'Ground') &
    (poke ['Legendary'] == False)
].groupby (by=['Generation'])[['Type 1']].value_counts ()
# Dari hasil grouping & aggregating, jenis pokemon ground terbanyak ada pada pokemon generasi ke - 1
```

```python
poke [(poke ['Generation'] >= 1) &
    (poke ['Type 1'] == 'Water') &
    (poke ['Legendary'] == False)
].groupby (by=['Generation'])[['Type 1']].value_counts ()
# Dari hasil grouping & aggregating, jenis pokemon water terbanyak ada pada pokemon generasi ke - 1
```

```python
poke [(poke ['Generation'] >= 1) &
    (poke ['Type 1'] == 'Flying') &
    (poke ['Legendary'] == False)
].groupby (by=['Generation'])[['Type 1']].value_counts ()
# Dari hasil grouping & aggregating, jenis pokemon flying terbanyak ada pada pokemon generasi ke - 6
```
